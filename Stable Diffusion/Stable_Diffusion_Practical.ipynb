## Stable Diffusion
  It is a deep learning text-to-image model released in 2022. It generates detailed images from text descriptions, and can also be used for tasks like inpainting, outpainting, and image-to-image translation.


## Key components:

- Text encoder
- Latent diffusion model
- Image decoder

## The Diffusion Process

1. Start with random noise

- The process begins with a completely random arrangement of pixels, essentially pure noise with no recognizable picture or pattern.


2. Gradually denoise the image

- The model iteratively refines this noise, slowly transforming
  it into a coherent image by removing noise in small steps.

3. Use text prompt to guide the denoising process

- The text description provided by the user acts as a guide,
  influencing how the noise is removed to shape the final image according to the prompt.

```python
# stable diffusion model

import torch
from diffusers import StableDiffusionPipeline
```

```python
## Load the model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dytpe = torch.float16)
pipe = pipe.to("cuda")
```

```python
# Generate an image

prompt = "A serene landscape with moutanins and a lake"
image = pipe(prompt).images[0]


# Save tge image

image.save("generated_image.png")
```

```python
import torch
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image

model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Load an initial image
init_image = Image.open("/content/generated_image.png").convert("RGB")
init_image = init_image.resize((512, 512))

prompt = "A beautiful painting of a landscape"
images = pipe(prompt=prompt, image=init_image, strength=0.75, guidance_scale=7.5).images

images[0].save("output_image.png")
```

```python
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler

model_id = "stabilityai/stable-diffusion-2-1"

# Use the DPMSolverMultistepScheduler (DPM-Solver++) scheduler here instead
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")

prompt = "a photo of an astronaut riding a horse on mars"
image = pipe(prompt).images[0]

image.save("astronaut_rides_horse.png")
```
